# IRIS CLASSIFICATION
**First import all the needed packages for developing a model.**

```python
import numpy as np 
import pandas as pd 
from matplotlib import pyplot as plt
import seaborn as sns
```

**We will now read the dataset. I downloaded the data from the Kaggle.**

```python
data=pd.read_csv('IRIS.csv')
data
```

**After getting the dataset we will now get the details of the dataset.**

```python
data.info()
```

**Now we will check wether there are any null values in the dataset**

```python
data.isna()
```

**As there are no null values we will now directly divide the dataset**

```python
x=data.iloc[:,0:4]
x
```

```python
y=data.iloc[:,4:]
y
```

**Now we will split and the train the model using the dataset**

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.25, random_state=35)
xtrain
```

```python
xtest
```

```python
ytrain
```

```python
ytest
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(criterion='entropy',random_state=35)
dt
```

```python
dt.fit(xtrain,ytrain)
```

```python
from sklearn import tree
text_representation = tree.export_text(dt)
print(text_representation)
```

```python
fig = plt.figure(figsize = (25 , 20) , dpi = 200.0)

_ = tree.plot_tree(dt, 
                   feature_names = ['sepal.length' ,'sepal.width', 'petal.length', 'petal.width'],
                   class_names = ['setosa', 'versicolor', 'virginica'],
                   filled = True)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
pred=dt.predict(xtest)
accuracy=accuracy_score(ytest,pred)*100
accuracy
```

```python
from sklearn.svm import SVC
sv=SVC()
sv.fit(xtrain,ytrain)
```

```python
predy=sv.predict(xtest)
acc=accuracy_score(ytest,predy)
acc
```

```python
from sklearn import neighbors
clas=neighbors.KNeighborsClassifier()
clas
```

```python
clas.fit(xtrain,ytrain)
prediction=clas.predict(xtest)
accuracy=accuracy_score(ytest,prediction)
accuracy*100
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=32)
rf
```

```python
rf.fit(xtrain,ytrain)
p=rf.predict(xtest)
a=accuracy_score(ytest,p)*100
a
```

**Applying the Decision Tree model to classify a new sample.**

```python
# Use the decision tree model to predict class label of new sample i.e., classify new sample.

# creating new iris flower's data and loading into Dataframe

new_data = {'sepal_length' : [3.7], 
            'sepal_width' :  [3.0], 
            'petal_length' : [2.2], 
            'petal_width' : [1.3] }

new_df = pd.DataFrame(new_data)

new_df.head()
```

```python
new_pred = dt.predict(new_df)

print('Prediction: The variety of the iris flower is ' , new_pred)
```
